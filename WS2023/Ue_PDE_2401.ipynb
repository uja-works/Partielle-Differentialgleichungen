LÃ¶sen der PDE 
$$ u_t + x u_x = 0.$$

Visualisieren des Tangentialvektorfeldes.

```python
NX = NY = 30
xx = np.linspace(-1, 1, NX)
yy = np.linspace(-1, 1, NY)

X, Y = np.meshgrid(xx, yy)
U = 1
V = X
plt.figure()
plt.quiver(X, Y, U, V)
plt.xlabel("t")
plt.ylabel("x")
plt.axis("equal")
plt.plot()
```

## Analytische LÃ¶sung der PDE mit Sympy

```python
from sympy import Function, Eq
from sympy.solvers.pde import pdsolve
t, x = sym.symbols("t x")
u = Function("u")(t, x)
u_t = u.diff(t)
u_x = u.diff(x)
eq = Eq(u_t + x*u_x, 0)
pdsolve(eq)
```

Wie sieht es aus, wenn wir die Reihenfolge der Variablen vertauschen?

```python
from sympy import Function, Eq
from sympy.solvers.pde import pdsolve
t, x = sym.symbols("t x")
#u = Function("u")(x, t)
# Alternative Definition:
u = sym.symbols("u", cls=Function)
u = u(x,t)
u_t = u.diff(t)
u_x = u.diff(x)
eq = Eq(u_t + x*u_x, 0)
pdsolve(eq).simplify()
```

Beide LÃ¶sungen sind korrekt und gehen durch Verkettung von $F$ mit der log- bzw. exp-Funktion auseinander hervor.
