## Orthogonalität der Sinusfunktionen

```python
pi = sym.pi
L, x = sym.symbols("L x")
m, n = sym.symbols("m n", integer=True, positive=True)
```

```python
sym.integrate(sym.sin(m*pi*x/L) * sym.sin(n*pi*x/L), (x, 0, L)).simplify()
```

```python
m, n = sym.symbols("m n", integer=True, positive=True, odd=True)
sym.integrate(sym.sin(m*pi*x/(2*L)) * sym.sin(n*pi*x/(2*L)), (x, 0, L)).simplify()
```

## Lösung von partiellen Differentialgleichungen erster Ordnung

```python
from sympy import Function, Eq
from sympy.solvers.pde import pdsolve
t, x = sym.symbols("t x")
v = sym.symbols("v")
u = Function("u")(t, x)
u_t = u.diff(t)
u_x = u.diff(x)
eq = Eq(u_t + v*x*u_x, 0)
pdsolve(eq)
```

Warnung: bei Umbenennung der Variablen können andere, äquivalente Lösungen herauskommen.

```python
from sympy import Function, Eq
from sympy.solvers.pde import pdsolve
t, x = sym.symbols("t x")
v = sym.symbols("v")
u = Function("u")(t, x)
u_t = u.diff(t)
u_x = u.diff(x)
eq = Eq(v*t*u_t + u_x, 0)
pdsolve(eq)
```
